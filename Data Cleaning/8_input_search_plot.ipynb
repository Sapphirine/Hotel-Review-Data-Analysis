```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import folium
%matplotlib inline

data3 = pd.read_csv("hotel_review_d.csv")
```

```python
key1 = input("What is your first key word? ")
if "business"==key1:
    print("~~~")
```

```python
#data3 = pd.read_csv("hotel_review_di.csv")
from folium.plugins import MarkerCluster
import re
from folium import plugins

key1 = input("What is your first key word? (deluxe)")
key2 = input("What is your second key word? (business)")
key3 = input("What is your third key word? (comfort)")
m = folium.Map(location=[52.36,4.91], zoom_start=5)

marker_cluster = MarkerCluster().add_to(m)

for i in range(30):
    lat=data3['lat'][i]
    lng=data3['lng'][i]
    # Use regular expressions to do a find-and-replace
    letters_only = re.sub("[^a-zA-Z]",           # The pattern to search for
                      " ",                   # The pattern to replace it with
                      data3['Tags'][i] )  # The text to search
    lower_case = letters_only.lower()        # Convert to lower case
    words = lower_case.split()               # Split into words
    num=len(words)
    j=0
    for w in words:
        j += 1
        if w == key1:
            folium.Marker(        
                location=[ lat, lng],
                popup=str(data3['Hotel_Name'][i]),
                icon=folium.Icon(color='green', icon='music'),
            ).add_to(marker_cluster)
            #print ("y")
            break
        elif w == key2:
            #print ("n")
            folium.Marker(        
                location=[ lat, lng],
                popup=str(data3['Hotel_Name'][i]),
                icon=folium.Icon(color='red', icon='book'),
            ).add_to(marker_cluster)
            break
        elif w == key3:
            #print ("n")
            folium.Marker(        
                location=[ lat, lng],
                popup=str(data3['Hotel_Name'][i]),
                icon=folium.Icon(color='blue', icon='home'),
            ).add_to(marker_cluster)
            break
        elif j == num:
            #print ("n")
            #print ("n")
            A=str(data3['Hotel_Name'][i])
            B=str(data3['Average_Score'][i])
            C=str(data3['Total_Number_of_Reviews'][i])
            c_name=['Hotel_Name', 'Avg_Score','#Reviews']
            df = pd.DataFrame(data=[[A, B, C]], columns=c_name)
            html = df.to_html(classes='table table-striped \
                              table-hover table-condensed table-responsive')
            folium.Marker(        
                location=[ lat, lng],
                popup=html
                #popup=str(data3['Hotel_Name'][i])+'\n'+str(data3['Average_Score'][i]),
                #icon=folium.Icon(color='blue', icon='ok-sign'),
            ).add_to(marker_cluster)
        #print(j)

#hotel_names=data3['Hotel_Name'][:30]
#plugins.Search(hotel_names, search_zoom=6).add_to(m)

m.save('30_inputsearch_Marker_HR_di.html')

m
```
