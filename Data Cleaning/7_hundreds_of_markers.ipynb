```python
from folium.plugins import MarkerCluster
import re
from folium import plugins
import folium
```

```python
import pandas as pd
data = pd.read_csv("hotel_review_di.csv")
data.head(3)
```

```python
m = folium.Map(location=[52.36,4.91], zoom_start=5)
marker_cluster = MarkerCluster().add_to(m)

data = pd.read_csv("hotel+tag2.csv")
data3 = pd.read_csv("hotel_review_di.csv")
data3 = data3[pd.notnull(data3["lat"])]
data3 = data3[pd.notnull(data3["lng"])]
for i in range(1000):
    lat=data3['lat'][i]
    lng=data3['lng'][i]
    # Use regular expressions to do a find-and-replace
    letters_only = re.sub("[^a-zA-Z]",           # The pattern to search for
                      " ",                   # The pattern to replace it with
                      data['key'][i] )  # The text to search
    lower_case = letters_only.lower()        # Convert to lower case
    words = lower_case.split()               # Split into words
    num=len(words)
    j=0
    for w in words:
        j += 1
        if w == "clean":
            folium.Marker(        
                location=[ lat, lng],
                popup=str(data3['Hotel_Name'][i]),
                icon=folium.Icon(color='blue', icon='cloud'),
            ).add_to(marker_cluster)
            #print ("y")
            break
        elif w == "breakfast":
            #print ("n")
            folium.Marker(        
                location=[ lat, lng],
                popup=str(data3['Hotel_Name'][i]),
                icon=folium.Icon(color='green', icon='glass'),
            ).add_to(marker_cluster)
            break
        elif w == "beautiful":
            #print ("n")
            folium.Marker(        
                location=[ lat, lng],
                popup=str(data3['Hotel_Name'][i]),
                icon=folium.Icon(color='beige', icon='ok-sign'),
            ).add_to(marker_cluster)
            break
        elif j == num:
            #print ("n")
            A=str(data3['Hotel_Name'][i])
            B=str(data3['Average_Score'][i])
            C=str(data3['Total_Number_of_Reviews'][i])
            c_name=['Hotel_Name', 'Avg_Score','#Reviews']
            df = pd.DataFrame(data=[[A, B, C]], columns=c_name)
            html = df.to_html(classes='table table-striped \
                              table-hover table-condensed table-responsive')
            folium.Marker(        
                location=[ lat, lng],
                popup=html
                #popup=str(data3['Hotel_Name'][i])+'\n'+str(data3['Average_Score'][i]),
                #icon=folium.Icon(color='blue', icon='ok-sign'),
            ).add_to(marker_cluster)
        #print(j)

hotel_names=data['Hotel_Name'][:10]
#plugins.Search(hotel_names, search_zoom=2).add_to(m)

m.save('1000_markers_12_12_17.html')

m
```
